```python
# -*- coding:utf-8 -*-

from __future__ import division ## pour la division

import sys    
import os
import codecs

import pandas as pd
import matplotlib.pyplot as plt
import numpy

###################################################################
###  Lecture du fichier en format csv 

data = pd.read_csv("trafic-de-voyageurs-et-marchandises-depuis-1841.csv",
                   sep = ";",encoding="UTF-8")

## liste pour encoding : https://docs.python.org/3/library/codecs.html#standard-encodings
nb_Lignes=len(data)
nb_Colonnes=len(data.columns)

print("nombre de lignes =%d"%len(data))

print("nombre de colonnes =%d"%len(data.columns))

print("Noms des colonnes : %s"%data.columns)

## On ordonne les données par ordre croissant selon les
# valeur des la 1ère Colonne (année)
data=data.sort_values(by=[data.columns[0]], ascending=[True])

Noms=[data.columns[i] for i in range(5)]  # noms des colonnes 

# extraire des données , uniqueemnt les colonnes non vides 1,...,5
#data1 = pd.DataFrame(data,columns=Noms , dtype='float64')

data1 = pd.DataFrame(data,columns=Noms)

# Transformer les données : col2,...,col5 en numérqie
for i in range(1,5):
      data1[data1.columns[i]] =pd.to_numeric(data1[data1.columns[i]])

# indexation des données selon année : data1.columns[0]
data2 = data1.set_index(data.columns[0])

#print(data2.columns)
#print(data2[data2.columns[0]]) # 1ère colonne

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111)
# tracer la 2ème colonne
plt.plot(data2[data2.columns[0]], lw=2)

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.set_xlabel("{}".format(data.columns[0]), labelpad=5, fontsize=14, fontname='serif', color="blue")
#ax.set_ylabel("nombre de {}".format(data.columns[1]), labelpad=5, fontsize=14, fontname='serif', color="blue")
ax.set_title("Nombre de {0} \n entre {1} et {2}".format(data.columns[1],int(data2.index[0]),int(data2.index[len(data2)-1])),
            fontsize=14,fontname='serif', color="blue")

plt.savefig('Figure001.eps', format="eps", dpi=400)
#plt.savefig('Figure001.png', format="png", dpi=400)
plt.show()
plt.close()


################ récupérer le nom du fichier courant   ################
#print(os.path.basename(sys.argv[0]))
#print(sys.argv[0][:-3])

#######
Fichier_courant=os.path.splitext(os.path.basename(sys.argv[0]))
#### --->  ('code_001', '.py')
#######
Fichier_resultats=Fichier_courant[0]+"_stats.txt"

#print(Fichier_resultats)

pd.set_option('display.precision', 4)
temp =data2[data2.columns[0]].describe()[['count','min','max','mean','std','25%','50%','75%']]


f = codecs.open(Fichier_resultats, mode="w", encoding="utf-8")
f.write("----------------------------------------------------------- \n")
f.write("\t\t statistique descriptive des données\t \t\n")
f.write("----------------------------------------------------------- \n")
f.write(temp.to_string())
f.write(" \n")
f.write("----------------------------------------------------------- \n")
f.close()
```

<h3>  code_1_loi_normal.py </h3>

```python
import numpy
import scipy
import scipy.stats

######   fonction randn de numpy  ######

#v.a. N(0,1)
numpy.random.randn()
# vecteur de dim=2 suivant la loi  N(0,1)
numpy.random.randn(2,1)
# matrice de 5-lignes, 2-colonnes suivant la loi  N(0,1)
numpy.random.randn(5,2)
#v.a. N(2,5)
2.0+numpy.sqrt(5)*numpy.random.randn()
# matrice de 3-lignes, 5-colonnes suivant la loi  N(-2.4,11)
-2.4+numpy.sqrt(11)*numpy.random.randn(3,5)

######   fonction numpy.random.normal   ######
#v.a. N(0,1)
numpy.random.normal()
#v.a. N(8,12)
numpy.random.normal(8,numpy.sqrt(12))
## matrice de 7-lignes, 3-colonnes suivant la loi  N(4,6)
numpy.random.normal(4,numpy.sqrt(6),[7,3])

######    fonction scipy.stats.norm  : import scipy.stats  ######

#v.a. N(0,1)
scipy.stats.norm.rvs(loc=0, scale=1, size=1)
## matrice de 8-lignes, 4-colonnes suivant la loi  N(-3,2)
scipy.stats.norm.rvs(loc=-3.0, scale=numpy.sqrt(2), size=[8,4])
```

<h3>  code_2 _ histogramme_normale.py </h3>

```python

from matplotlib import pylab as plt
import numpy

#### fonction pour enlever les cardres haut , gauche droit de la figure
def Axe_simple(ax):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.set_yticks([])

### simulation de la loi normale N(mu,sigma2)
taille =7000
mu=5
sigma2=7
data=numpy.random.normal(mu, numpy.sqrt(sigma2), size=taille)
#data : vecteur de taille=2000

### histogramme
nb_classe=15
min_x=numpy.min(data)
max_x=numpy.max(data)

#extremum_classe : extrémités des classes pour l'histogramme
extremum_classe=[min_x+(max_x-min_x)*i/(nb_classe) for i in range(nb_classe+1)]

fig, ax_h = plt.subplots(1, 1)
fig.set_size_inches(16.5, 10.5)
ax_h.hist(data,extremum_classe,facecolor=[0.50,0.50,0.80], edgecolor='white')
ax_h.set_title("simulation d'un échantillon de taill %d de la loi $\mathcal{N}(%s,%s)$"%(taille,mu,sigma2))

Axe_simple(ax_h)
plt.show()
```

<h3> code_4_ processus_aleatoire_temps_continu.py </h3>

```python
from matplotlib import pylab as plt
import math
import numpy


def Processus_Continu(t,n=100,  sigma2=1):
    m=math.floor(n*t) # [nt]
    s=numpy.sqrt(sigma2)
    return sum([numpy.random.normal(0,s) for _ in range(m)])

##plt.figure(figsize=(10, 6), dpi=250)
plt.figure(figsize=(10, 6))
ax = plt.gca()

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.xaxis.set_ticks_position('bottom')

nb_trajectoire=[100, 1000, 1500, 5000, 6000]
nb_point=500
t=[0.01+(0.99-0.01)*i/(nb_point-1)  for i in range(nb_point)]
for j in range(len(nb_trajectoire)):
    ax.plot(t, [Processus_Continu(temps,n=nb_trajectoire[j],sigma2=1) for temps in t])
ax.axhline(y=0,color='black',lw=1)

ax.set_xlabel("t")
ax.set_ylabel("$X_{t}$")
ax.set_title("{:} trajectoires du processus".format(len(nb_trajectoire)))

##plt.savefig('Processus_alea.eps', format="eps", dpi=400)
##plt.savefig('Processus_alea.png', format="png", dpi=400)
plt.show()
####plt.show() avant plt.close()
##plt.close()
```
